```python
# Force-to-Bodyweight Ratio Calculator
# Used in vertical jump and strength assessments

# Input data
athlete_name = "Test Athlete"
body_weight = 750  # Newtons (mass × 9.81)
peak_force = 2250  # Newtons from force plate

# Calculate ratio
force_ratio = peak_force / body_weight

# Interpretation
print("🏃 FORCE ANALYSIS")
print("=" * 40)
print(f"Athlete: {athlete_name}")
print(f"Body Weight: {body_weight} N")
print(f"Peak Force: {peak_force} N")
print(f"Force-to-BW Ratio: {force_ratio:.2f}")
print()
print("Interpretation:")
if force_ratio >= 3.0:
    print("✅ Elite level (3.0+)")
elif force_ratio >= 2.5:
    print("✅ Advanced (2.5-3.0)")
elif force_ratio >= 2.0:
    print("⚠️ Intermediate (2.0-2.5)")
else:
    print("⚠️ Needs development (<2.0)")
```
