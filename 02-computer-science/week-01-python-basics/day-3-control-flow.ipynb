```python
# IF/ELSE - Two possible outcomes

heart_rate = 175
print("Heart Rate Analysis")
print("=" * 30)
print(f" Current Heart Rate: {heart_rate} bpm")

if heart_rate > 180:
    print ("High Intensity Zone")
    print ("Consider reducing effort")
else:
    print ("Safe Zone")
    print ("You Can Maintain This Pace")
```

```python
# IF/ELIF/ELSE - Multiple heart rate zones

heart_rate = 165

print("🏃 TRAINING ZONE ANALYSIS")
print("=" * 40)
print(f"Current HR: {heart_rate} bpm")
print()

if heart_rate >= 180:
    print("🔴 ZONE 5: Maximum Effort")
    print("→ 90-100% max HR")
    print("→ Only sustainable for short bursts")
elif heart_rate >= 160:
    print("🟠 ZONE 4: Hard Effort") 
    print("→ 80-90% max HR")
    print("→ Lactate threshold training")
elif heart_rate >= 140:
    print("🟡 ZONE 3: Moderate Effort")
    print("→ 70-80% max HR")
    print("→ Aerobic endurance building")
elif heart_rate >= 120:
    print("🟢 ZONE 2: Light Effort")
    print("→ 60-70% max HR")
    print("→ Fat burning zone")
else:
    print("⚪ ZONE 1: Very Light / Recovery")
    print("→ 50-60% max HR")
    print("→ Active recovery")

print()
print("💡 Recommendation: Adjust intensity based on training goal")
```

```python
# Vertical Jump Performance Assessment
# Based on force-to-bodyweight ratio

print("🏀 VERTICAL JUMP ASSESSMENT")
print("=" * 50)

# Input data
athlete_name = "Bruno"
body_mass = 85  # kg
peak_force = 2550  # Newtons from force plate

# Calculate
body_weight = body_mass * 9.81  # Convert to Newtons
force_ratio = peak_force / body_weight

# Display measurements
print(f"\nAthlete: {athlete_name}")
print(f"Body Mass: {body_mass} kg")
print(f"Body Weight: {body_weight:.1f} N")
print(f"Peak Force: {peak_force} N")
print(f"Force-to-BW Ratio: {force_ratio:.2f}")

# Performance classification
print("\n📊 PERFORMANCE CLASSIFICATION:")
print("-" * 50)

if force_ratio >= 3.5:
    print("⭐⭐⭐ ELITE LEVEL (3.5+)")
    print("→ Professional/Olympic athlete range")
    print("→ Excellent explosive power")
    recommendation = "Maintain strength, focus on technique refinement"
elif force_ratio >= 3.0:
    print("⭐⭐ ADVANCED LEVEL (3.0-3.5)")
    print("→ High-level collegiate/competitive athlete")
    print("→ Strong explosive power")
    recommendation = "Continue power training, add plyometrics"
elif force_ratio >= 2.5:
    print("⭐ INTERMEDIATE LEVEL (2.5-3.0)")
    print("→ Recreational athlete/developing competitor")
    print("→ Good foundation")
    recommendation = "Focus on strength training 3x/week"
elif force_ratio >= 2.0:
    print("📈 DEVELOPING (2.0-2.5)")
    print("→ Beginning athlete")
    print("→ Room for significant improvement")
    recommendation = "Build strength foundation, squat/deadlift focus"
else:
    print("🎯 BEGINNER (<2.0)")
    print("→ New to athletic training")
    print("→ Significant potential for gains")
    recommendation = "Start with bodyweight exercises, progress to weights"

print(f"\n💡 RECOMMENDATION:")
print(f"   {recommendation}")

# Training readiness check
print(f"\n🔥 TRAINING READINESS:")
if force_ratio >= 2.5:
    print("   ✅ Ready for high-intensity plyometrics")
else:
    print("   ⚠️ Build strength base before advanced plyometrics")
```
