```python
# Complete Team Force Analysis - Real Research Tool

print("🏋️ TEAM FORCE PLATE ANALYSIS")
print("=" * 60)

# Team data (from force plate testing)
athletes = ["Alex", "Bruno", "Carlos", "Diana", "Elena"]
body_mass = [75, 85, 70, 65, 80]  # kg
peak_force = [2100, 2550, 1950, 2200, 2400]  # Newtons

print(f"Analyzing {len(athletes)} athletes...")
print()

# Analyze each athlete
for i in range(len(athletes)):
    name = athletes[i]
    mass = body_mass[i]
    force = peak_force[i]
    
    # Calculate
    body_weight = mass * 9.81  # Convert to Newtons
    force_ratio = force / body_weight
    
    # Display individual results
    print(f"📊 {name}:")
    print(f"   Body mass: {mass} kg")
    print(f"   Peak force: {force} N")
    print(f"   Force-to-BW ratio: {force_ratio:.2f}")
    
    # Performance classification
    if force_ratio >= 3.5:
        classification = "⭐⭐⭐ ELITE"
        recommendation = "Maintain strength, refine technique"
    elif force_ratio >= 3.0:
        classification = "⭐⭐ ADVANCED"
        recommendation = "Add plyometrics"
    elif force_ratio >= 2.5:
        classification = "⭐ INTERMEDIATE"
        recommendation = "Strength training 3x/week"
    else:
        classification = "📈 DEVELOPING"
        recommendation = "Build strength foundation"
    
    print(f"   {classification}")
    print(f"   → {recommendation}")
    print()

# Team summary statistics
print("=" * 60)
print("📈 TEAM SUMMARY:")

# Calculate team averages
avg_mass = sum(body_mass) / len(body_mass)
avg_force = sum(peak_force) / len(peak_force)

# Calculate all ratios
all_ratios = []
for i in range(len(athletes)):
    ratio = peak_force[i] / (body_mass[i] * 9.81)
    all_ratios.append(ratio)

avg_ratio = sum(all_ratios) / len(all_ratios)
best_ratio = max(all_ratios)
worst_ratio = min(all_ratios)

print(f"   Average body mass: {avg_mass:.1f} kg")
print(f"   Average peak force: {avg_force:.1f} N")
print(f"   Average force ratio: {avg_ratio:.2f}")
print(f"   Best ratio: {best_ratio:.2f}")
print(f"   Worst ratio: {worst_ratio:.2f}")
```

```python
# Function That Classifies Performance

def classify_jump_performance(jump_height):
    """
    Classify vertical jump performance
    Input: jump height in meters
    Output: classification string
    """
    if jump_height >= 0.70:
        return "🏆 ELITE (70+ cm)"
    elif jump_height >= 0.60:
        return "⭐⭐⭐ EXCELLENT (60-70 cm)"
    elif jump_height >= 0.50:
        return "⭐⭐ GOOD (50-60 cm)"
    elif jump_height >= 0.40:
        return "⭐ AVERAGE (40-50 cm)"
    else:
        return "📈 NEEDS WORK (<40 cm)"

# Test the function
print("🏀 JUMP PERFORMANCE CLASSIFIER")
print("=" * 50)

athletes = ["Alex", "Bruno", "Carlos", "Diana", "Elena"]
jump_heights = [0.45, 0.65, 0.38, 0.72, 0.58]

for i in range(len(athletes)):
    name = athletes[i]
    height = jump_heights[i]
    classification = classify_jump_performance(height)
    
    print(f"{name}: {height}m → {classification}")
```

```python
# Function Returning Multiple Values

def analyze_athlete(body_mass, peak_force):
    """
    Complete athlete analysis
    Returns: body_weight, force_ratio, classification
    """
    # Calculate
    body_weight = body_mass * 9.81
    force_ratio = peak_force / body_weight
    
    # Classify
    if force_ratio >= 3.5:
        classification = "ELITE"
    elif force_ratio >= 3.0:
        classification = "ADVANCED"
    elif force_ratio >= 2.5:
        classification = "INTERMEDIATE"
    else:
        classification = "DEVELOPING"
    
    # Return THREE values
    return body_weight, force_ratio, classification


# Use the function
print("🏋️ COMPLETE ATHLETE ANALYSIS")
print("=" * 60)

athletes = ["Alex", "Bruno", "Carlos"]
masses = [75, 85, 70]
forces = [2100, 2550, 1950]

for i in range(len(athletes)):
    name = athletes[i]
    mass = masses[i]
    force = forces[i]
    
    # Get all three values back!
    bw, ratio, level = analyze_athlete(mass, force)
    
    print(f"\n{name}:")
    print(f"  Body weight: {bw:.1f} N")
    print(f"  Force ratio: {ratio:.2f}")
    print(f"  Level: {level}")
```

```python
# ========================================
# WEEK 1 CAPSTONE PROJECT
# Team Performance Analysis System
# Combines: Variables, Lists, Loops, Functions, Conditionals
# ========================================

# === FUNCTIONS ===

def calculate_bmi(mass, height):
    """Calculate BMI"""
    return mass / (height ** 2)

def calculate_force_ratio(mass, force):
    """Calculate force-to-bodyweight ratio"""
    body_weight = mass * 9.81
    return force / body_weight

def classify_performance(ratio):
    """Classify athlete performance level"""
    if ratio >= 3.5:
        return "ELITE", "⭐⭐⭐"
    elif ratio >= 3.0:
        return "ADVANCED", "⭐⭐"
    elif ratio >= 2.5:
        return "INTERMEDIATE", "⭐"
    else:
        return "DEVELOPING", "📈"


# === TEAM DATA ===

team_name = "WSU Biomechanics Lab"
athletes = ["Alex", "Bruno", "Carlos", "Diana", "Elena"]
body_mass = [75, 85, 70, 65, 80]  # kg
heights = [1.80, 1.77, 1.75, 1.68, 1.82]  # meters
peak_forces = [2100, 2550, 1950, 2200, 2400]  # Newtons


# === MAIN ANALYSIS ===

print("=" * 70)
print(f"🏋️  {team_name.upper()} - PERFORMANCE REPORT")
print("=" * 70)
print(f"\nTotal Athletes: {len(athletes)}")
print()

# Store results for summary
all_ratios = []
all_bmis = []

# Analyze each athlete
for i in range(len(athletes)):
    name = athletes[i]
    mass = body_mass[i]
    height = heights[i]
    force = peak_forces[i]
    
    # Calculate metrics
    bmi = calculate_bmi(mass, height)
    ratio = calculate_force_ratio(mass, force)
    level, stars = classify_performance(ratio)
    
    # Store for summary
    all_ratios.append(ratio)
    all_bmis.append(bmi)
    
    # Display individual report
    print(f"{'─' * 70}")
    print(f"ATHLETE: {name}")
    print(f"{'─' * 70}")
    print(f"  Body Mass:        {mass} kg")
    print(f"  Height:           {height} m")
    print(f"  BMI:              {bmi:.1f}")
    print(f"  Peak Force:       {force} N")
    print(f"  Force/BW Ratio:   {ratio:.2f}")
    print(f"  Classification:   {stars} {level}")
    
    # Recommendations
    print(f"  Recommendation:   ", end="")
    if level == "ELITE":
        print("Maintain strength, refine technique")
    elif level == "ADVANCED":
        print("Add plyometric training")
    elif level == "INTERMEDIATE":
        print("Increase strength training frequency")
    else:
        print("Build foundational strength")
    print()


# === TEAM SUMMARY ===

print("=" * 70)
print("📊 TEAM SUMMARY STATISTICS")
print("=" * 70)

avg_mass = sum(body_mass) / len(body_mass)
avg_force = sum(peak_forces) / len(peak_forces)
avg_ratio = sum(all_ratios) / len(all_ratios)
avg_bmi = sum(all_bmis) / len(all_bmis)

print(f"\n  Average Body Mass:    {avg_mass:.1f} kg")
print(f"  Average Peak Force:   {avg_force:.1f} N")
print(f"  Average Force Ratio:  {avg_ratio:.2f}")
print(f"  Average BMI:          {avg_bmi:.1f}")

print(f"\n  Best Force Ratio:     {max(all_ratios):.2f} ({athletes[all_ratios.index(max(all_ratios))]})")
print(f"  Lowest Force Ratio:   {min(all_ratios):.2f} ({athletes[all_ratios.index(min(all_ratios))]})")

# Count performance levels
elite_count = sum(1 for r in all_ratios if r >= 3.5)
advanced_count = sum(1 for r in all_ratios if 3.0 <= r < 3.5)
intermediate_count = sum(1 for r in all_ratios if 2.5 <= r < 3.0)
developing_count = sum(1 for r in all_ratios if r < 2.5)

print(f"\n  Performance Distribution:")
print(f"    ⭐⭐⭐ Elite:         {elite_count}")
print(f"    ⭐⭐ Advanced:       {advanced_count}")
print(f"    ⭐ Intermediate:    {intermediate_count}")
print(f"    📈 Developing:      {developing_count}")

print("\n" + "=" * 70)
print("✅ ANALYSIS COMPLETE")
print("=" * 70)
```
